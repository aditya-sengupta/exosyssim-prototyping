```python
# let's group all these planets into buckets based on planet size and orbital period!
radii = np.array(planetary['koi_prad'])
plt.hist(np.log(radii), bins=100)
```

```python
periods = np.array(planetary['koi_period'])
plt.hist(np.log(periods), bins=100)
```

```python
fposes = np.array(planetary['koi_disposition']) == 'CONFIRMED'
print(fposes)
```

```python
plt.scatter(np.log(periods), np.log(radii))
plt.xlabel("Log-period (ln days)")
plt.ylabel("Log-radius (ln Earth radii)")
```

```python
min_period, min_radius = min(periods), min(radii)
max_period, max_radius = max(periods), max(radii)
period_buckets = [min_period * 2 ** i for i in range(int(np.ceil(np.log2(max_period / min_period))))]
radius_buckets = [min_radius * 2 ** i for i in range(int(np.ceil(np.log2(max_radius / min_radius))))]
hits = np.zeros((len(period_buckets), len(radius_buckets)))
fpos = np.zeros((len(period_buckets), len(radius_buckets)))
for radius, period, conf in zip(radii, periods, fposes):
    if np.isfinite(radius) and np.isfinite(period):
        p_index = int(np.log2(period / min_period))
        r_index = int(np.log2(radius / min_radius))
        if conf:
            hits[p_index][r_index] += 1
        else:
            fpos[p_index][r_index] += 1
        
print("Hits: ", np.sum(hits))
print("Misses: ", np.sum(fpos))

plt.figure(figsize=(10,8))
plt.imshow(hits / np.sum(hits), origin='lower') # tbd: fix axes to match period_buckets and radius_buckets
plt.colorbar()
plt.xlabel("Planetary period ('bucket' units)")
plt.ylabel("Planetary radius ('bucket' units)")
plt.title("True planetary detections")
```

```python
plt.figure(figsize=(10,8))
plt.imshow(fpos / np.sum(fpos), origin='lower') # tbd: fix axes to match period_buckets and radius_buckets
plt.colorbar()
plt.xlabel("Planetary period ('bucket' units)")
plt.ylabel("Planetary radius ('bucket' units)")
plt.title("False postive detections")
```

```python
# make an array of inverse detection efficiencies: ratios of FP to True over each bucket.
# A lot of these are going to be 0 - maybe ABC will fix that!
ide = np.true_divide(fpos, hits, where=(np.logical_and(hits, fpos))!=0)
plt.imshow(np.round(ide), origin='lower')
plt.colorbar()
plt.xlabel("Planetary period ('bucket' units)")
plt.ylabel("Planetary radius ('bucket' units)")
plt.title("Inverse detection efficiencies")
```
